```python
# Importing all required packages

from app_store_scraper import AppStore

import pandas as pd

import numpy as np

import json

```

```python
# Defining required criterias for scraping

NatWest = AppStore(country='GB', app_name='NatWest Mobile Banking', app_id=334855322)

NatWest.review(how_many=3500)

NatWest.reviews



```

```python
# Converting scraped reviews to a dataframe 

df = pd.DataFrame(np.array(NatWest.reviews),columns=['review'])

df2 = df.join(pd.DataFrame(df.pop('review').tolist()))

df2.head()
```

```python
# Exporting reviews as a csv file
df2

df2.to_csv('NatWestAppStore2020rows.csv')
```
